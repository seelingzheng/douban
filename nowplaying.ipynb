```python
from requests.exceptions import RequestException
import requests
from bs4 import BeautifulSoup
import pandas as pd
BASE_CITY = 'beijing'
BASE_URL = 'https://movie.douban.com/cinema/nowplaying/{}/'



BASE_HEADER = {
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8",
    "Accept-Encoding": "gzip, deflate, br",
    "Accept-Language": "zh-CN,zh;q=0.9,en;q=0.8",
    "Connection": "keep-alive",
    "User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/63.0.3239.84 Safari/537.36"
}

CURRENT_HEADER = {
    'Cookie': 'll="108288"; bid=bfr8-Ko0YpU; _vwo_uuid_v2=9F7BE84EC3E3947C04E1385C0E57E4C3|5930c13d99e8d03d1a33ce50d169ef88; __yadk_uid=nRhcDZFePvTI3EgkTYAqAvIsq8lUoFZK; viewed="27599979"; _pk_ref.100001.4cf6=%5B%22%22%2C%22%22%2C1513862192%2C%22https%3A%2F%2Fwww.douban.com%2F%22%5D; _pk_ses.100001.4cf6=*; _pk_id.100001.4cf6=a8ba49c578010b06.1513341613.3.1513862322.1513525406.',
    'Host': 'movie.douban.com'
}


movielist = []

def get_now_playing(url):
    header = dict(CURRENT_HEADER, **BASE_HEADER)
    # print(header)
    try:
        res = requests.get(url, headers=header)
        if res.status_code == 200:
            soup = BeautifulSoup(res.text, 'lxml')
            lists = soup.select('#nowplaying > .mod-bd > .lists > .list-item')

            for item in lists:
                movielist.append( parse_item(item))
                

            return None
    except  RequestException:
        return None


def parse_item(html):
    arrtsKeys = html.attrs.keys()
    item = {}
    for key in arrtsKeys:
        if 'data-' in key:
            newKey = key.lstrip('data-')
            item[newKey] = html.attrs[key]

    item['link'] = html.li.a['href']
    item['poster']=html.li.a.img['src']
    return item
def main():
    url = BASE_URL.format(BASE_CITY)
    get_now_playing(url)

main()

df = pd.DataFrame(data=movielist,columns=['score','itle','ctors','irector','star','region'])
df.sort_values(['score'],ascending=False)  #排序

```
